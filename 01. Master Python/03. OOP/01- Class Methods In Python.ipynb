### Class


class is a blueprint for creating objects (a particular data structure), providing initial values for state (member variables or attributes), and implementations of behavior (member functions or methods).


```python

# example of class 
class Person:
    # this is a constructor 
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def myfunc(self):
        print("Hello my name is " + self.name)

p1 = Person("John", 36)
print(p1.name)
print(p1.age)
p1.myfunc()
```

### we can have multiple constractor

```python
# class with multiple constractors

class Person:
    def __init__(self, *args):
        if len(args)==1:
            self.name = args[0]
            self.age = 0
        elif len(args)==2:
            self.name = args[0]
            self.age = args[1]
        else:
            self.name = "Hassane"
            self.age = 0


    def myfunc(self):
        print("Hello my name is " + self.name)

p1 = Person("John", 36)

print(p1.name)
print(p1.age)
```

# Static methode

static method is a method that doesn't modify the object or class state. It takes neither a self nor a cls parameter (but of course it's free to accept an arbitrary number of other parameters).


```python
import datetime

class Car:
    base_price = 100000  # Class Variables

    def __init__(self, windows, doors, year):
        self.windows = windows
        self.doors = doors
        self.year = year

    def what_base_price(self):
        print("The base price is {}".format(Car.base_price))

    @classmethod
    def revise_base_price(cls, inflation):
        cls.base_price = cls.base_price + (cls.base_price * inflation)

    @staticmethod
    def check_year(year):
        return year == 2024

# Instance of Car
car1 = Car(4, 4, 2023)

# Check the year and adjust base price accordingly
if Car.check_year(car1.year):
    Car.revise_base_price(0.10)
    print(car1.base_price)
else:
    print("No Inflation")
```
