## data classes
the @dataclass decorator is used to create classes that are primarily used to store data
it is a decorator that adds special methods to the class
1. __init__ method is created automatically
2. __repr__ method is created automatically
3. __eq__ method is created automatically

```python
from dataclasses import dataclass

@dataclass
class Point:
    x: int
    y: int
    z: int

p1 = Point(1, 2, 3)
p2 = Point(1, 2, 3)
p3 = Point(1, 2, 4)

print(p1)

print(p1 == p2)

print(p1 == p3)


```

```python
# inheritance

@dataclass
class Point:
    x: int
    y: int

@dataclass
class Point3D(Point):
    z: int

p = Point3D(1, 2, 3)
print(p)
print(p.x)
```
