# Basic Arithmetic Operations

```python
a = 5
b = 3

# Addition
print('Addition:', a + b)
# Subtraction
print('Subtraction:', a - b)
# Multiplication
print('Multiplication:', a * b)
# Division
print('Division:', a / b)
# Floor Division
print('Floor Division:', a // b)
# Modulus
print('Modulus:', a % b)
# Exponentiation
print('Exponentiation:', a ** b)
```

## Control and Conditional Structures

```python
x = 20
if x < 30:
    print('x is less than 30')
elif x == 30:
    print('x is equal to 30')
else:
    print('x is greater than 30')
```

## Looping - For and While Loops

```python
# For Loop
for i in range(5):
    print(i)

# While Loop
j = 0
while j < 5:
    print(j)
    j += 1
```

## Typecasting

```python
int_to_float = float(10)
print(int_to_float)
float_to_int = int(10.5)
print(float_to_int)
num_to_string = str(15)
print(num_to_string)
string_to_num = int('20')
print(string_to_num)
```
