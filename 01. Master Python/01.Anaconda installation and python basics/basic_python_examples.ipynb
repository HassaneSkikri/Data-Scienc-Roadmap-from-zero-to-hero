# Basic Arithmetic Operations

```python
a = 5
b = 3

# Addition
print('Addition:', a + b)
# Subtraction
print('Subtraction:', a - b)
# Multiplication
print('Multiplication:', a * b)
# Division
print('Division:', a / b)
# Floor Division
print('Floor Division:', a // b)
# Modulus
print('Modulus:', a % b)
# Exponentiation
print('Exponentiation:', a ** b)
```

## Control and Conditional Structures

```python
x = 20
if x < 30:
    print('x is less than 30')
elif x == 30:
    print('x is equal to 30')
else:
    print('x is greater than 30')
```

## Looping - For and While Loops

```python
# For Loop
for i in range(5):
    print(i)

# While Loop
j = 0
while j < 5:
    print(j)
    j += 1
```

### == Vs Is

```python
# the diffrence between == and is operator in python is that == is used to compare the values of the objects and is is used to compare the identity of the objects

# example
a = [1, 2, 3]
b = [1, 2, 3]
print(a == b)  # True
print(a is b)  # False
```

### Shallow copy vs deep copy 

shallow copy is a bit-wise copy of an object. A new object is created that has an exact copy of the values in the original object. If any of the fields of the object are references to other objects, just the reference addresses are copied i.e., only the memory address is copied.

deep copy is a process in which the copying process occurs recursively. It means first constructing a new collection object and then, recursively, inserting copies into it of the objects found in the original.


```python
# example

import copy

list1 = [1, 2, [3, 5], 4]
list2 = copy.copy(list1)
list3 = copy.deepcopy(list1)

list1[2][0] = 7

print(list1)  # [1, 2, [7, 5], 4]

print(list2)  # [1, 2, [7, 5], 4]

print(list3)  # [1, 2, [3, 5], 4]

```
