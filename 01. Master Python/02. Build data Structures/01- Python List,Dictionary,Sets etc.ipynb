### Boolean Variables

- Boolean values are the two constant objects False and True. 
- They are used to represent truth values (other values can also be considered
false or true). 
- In numeric contexts (for example, when used as the argument to an
arithmetic operator), they behave like the integers 0 and 1, respectively.
- The built-in function bool() can be used to cast any value to a Boolean,
if the value can be interpreted as a truth value
- They are written as False and True, respectively.

# Functions in Python

Functions are reusable pieces of programs. They allow you to give a name to a block of statements, allowing you to run that block using the specified name anywhere in your program and any number of times. This is known as calling the function. We've already seen one example of a function, `print()`. But Python has many more functions, and you can define your own functions.

```python
def function_name(parameters):
    '''Docstring explaining function.'''
    # Function statements
    return [expression]

# Example of a function that calculates the square of a given number
def square(number):
    '''Returns the square of a number.'''
    return number ** 2

print(square(4))  # Outputs: 16
```

## Lambda Functions
A lambda function is a small anonymous function. A lambda function can take any number of arguments, but can only have one expression. The expression is evaluated and returned. Lambda functions are used when a simple function is used only once or for a short period in your code. It's less verbose than using a regular function.

```python
lambda arguments: expression

# Example of a lambda function that adds 10 to the number passed in as an argument
add_ten = lambda a : a + 10
print(add_ten(5))  # Outputs: 15

# Lambda function that multiplies argument a with argument b
multiply = lambda a, b : a * b
print(multiply(5, 6))  # Outputs: 30
```

## Iterators and Generators
Iterators are objects that can be iterated upon. Python uses the `__iter__()` method to return an iterator object of the class. The iterator object then uses the `__next__()` method to get the next item.
Generators are a simple way of creating iterators. A generator is a function that returns an object (iterator) which we can iterate over (one value at a time). It is as easy as defining a normal function with `yield` statement instead of a `return` statement.

```python
# Example of an iterator
my_list = [1, 3, 5, 7]
my_iter = iter(my_list)

print(next(my_iter))  # Outputs: 1
print(next(my_iter))  # Outputs: 3

# Example of a generator function
def my_gen(x, y):
    while x <= y:
        yield x
        x += 1

gen = my_gen(1, 5)
for i in gen:
    print(i)
```

## Exception Handling & Importing Libraries
Exception handling allows you to deal with unexpected errors in your Python programs. It can prevent your program from crashing if something unexpected occurs. Python uses `try` and `except` blocks to handle exceptions.

```python
# Example of exception handling
try:
    x = 1 / 0
except ZeroDivisionError:
    print('You cannot divide by zero!')

# Importing libraries
# You can import standard Python libraries using the `import` statement.
# Once imported, you can use functions and classes defined in those libraries.

import math
print(math.sqrt(25))  # Outputs: 5.0
```
