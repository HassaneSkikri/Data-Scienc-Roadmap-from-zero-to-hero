### Python Decorators

```python
# function copy
#   - function copy is a shallow copy of the function object itself
# closures
#  - a closure is a function object that remembers values in enclosing scopes even if they are not present in memory
# decorators
#  - a decorator is a function that takes another function and extends the behavior of the latter function without explicitly modifying it
```

```python
def welcome():
    return "Welcome to Python"
```

```python
wel=welcome()
del welcome
```

```python
wel
```

```python
welcome()
```

```python
### Closures

def main_welcome(msg):
   
    def sub_welcome_class():
        print("Welcome here")
        print(msg)
        print("Glad to have you here")
    return sub_welcome_class()

# the msg is remembered by the sub_welcome_class even if the main_welcome function is deleted 
```

```python
main_welcome("hassan")
```

```python
### Closures & initial Decorators
def main_welcome(func):
   
    def sub_welcome_class():
        print("Welcome here")
        print(func([1,2,3,4,5,6]))
        print("Glad to have you here")
    return sub_welcome_class()
```

```python
main_welcome(max)
```

```python
### Decorators
def main_welcome(func):
   
    def sub_welcome_class():
        print("Welcome here")
        func()
        print("Glad to have you here")
    return sub_welcome_class()
```

```python
def yourname():
    print("This is your name")
a=main_welcome(yourname)
```

```python
@main_welcome
def yourname():
    print("This is your name")
```
