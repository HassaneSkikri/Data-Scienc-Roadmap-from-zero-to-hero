## Mango Db with Python- Querying Json Documents

```python
import pymongo

client=pymongo.MongoClient('mongodb://127.0.0.1:27017/')

mydb=client['Employee']
empinfo=mydb.employeeinformation
```

```python
record={
        'firstname':'Ali',
        'lastname':'anwar',
        'department':'Analytics',
        'qualification':'BE',
        'age':29
        }
empinfo.insert_one(record)
```

```python
records=[{
        'firstname':'John',
        'lastname':'Doe',
        'department':'Analytics',
        'qualification':'statistics',
        'age':35
        
        },
         {
        'firstname':'John ',
        'lastname':'Smith',
        'department':'Analytics',
        'qualification':'masters',
        'age':30
        
        },
        {
        'firstname':'skikri',
        'lastname':'Hassane',
        'department':'Analytics',
        'qualification':'phd',
        'age':34
        
        },
        {
        'firstname':'Ram',
        'lastname':'anas',
        'department':'Analytics',
        'qualification':'master',
        'age':32
        
        }]
```

```python
empinfo.insert_many(records)
```

```python
## Simple way of querying
empinfo.find_one()
```

```python
## Select * from employeeinformation
for record in empinfo.find({}):
    print(record)
```

```python
## Query the json documents based on equality conditions
# Select * from employeeinformation where firstname=krish

for record in empinfo.find({'firstname':'Ali'}):
    print(record)
```

```python
## Query documents using query operators($in,$lt,$gt)
for record in empinfo.find({'qualification':{'$in':['phd','master']}}):
    print(record)
```

```python
## And and Query operators
for record in empinfo.find({'qualification':'master','age':{'$lt':35}}):
    print(record)
```

```python
## OR operators
for record in empinfo.find({'$or':[{'firstname':'Krish'},{'qualification':'BE'}]}):
    print(record)
```

```python
# exists operators

for record in empinfo.find({'qualification':{'$exists':True}}):
    print(record)
```

```python
## Update the records
empinfo.update_one({'firstname':'Ali'},{'$set':{'qualification':'phd'}})
for record in empinfo.find({'firstname':'Ali'}):
    print(record)
```

```python
## Delete the record
empinfo.delete_one({'firstname':'Ali'})
for record in empinfo.find({}):
    print(record)
```

```python
## projection in mongodb
for record in empinfo.find({},{'_id':0,'firstname':1,'lastname':1}):
    print(record)

## Limit the records
for record in empinfo.find({}).limit(2):
    print(record)
```
